This is python for making a chatbot system for answering queries about graduate program in Accenture UK. The data source can be referred to Links [1]

Program structure:
1. Sentence pre-processing
The program uses the pre-trained GloVe embedding matrix which converts words to words' meaning vectors. The embedding matrix was trained using corpus from Wikipedia 2014. For a proof of concept purpose, a relatively small embedding matrix at 400K vocab and 50d is used to save computation time. The matrice source can be downloaded from Links [2].

2. LSTM model training
LSTM is a deep learning techiqnue which includes forget gate, update gate and output gate. It enables the memory of preivous cell over the network. In this program dropout with rate 0.5 is used for regularisation purpose. 

3. Chatbot response
The chatbot will ask for user's name and make corresponding answer according to the prediction from LSTM model. The program will end when the user type "bye"

Links

[1] Accenture UK: https://www.accenture.com/gb-en/careers/graduate-faq

[2] GloVe: https://nlp.stanford.edu/projects/glove/

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import string 
#!/usr/bin/python
# -*- coding: UTF-8 -*-

from FAQ_function import *
```

```python
word_to_index,index_to_word, word_to_vec_map = read_glove_vecs('glove.6B.50d.txt')

data = pd.read_excel('FAQ_data.xlsx')
question = data['Question']
answer = data['Answer']
y_class = np.arange(answer.shape[0])

X_train = np.concatenate((question, answer))
Y_train = np.concatenate((y_class, y_class))
# X_test = question
# Y_test = np.arange(answer.shape[0])

# maxLen = len(max(X_train, key=len).split())
maxLen = 60

X_train_indices = sentences_to_indices(X_train, word_to_index, maxLen)
Y_train_oh = np.eye(answer.shape[0]+1)[Y_train.reshape(-1)]
# X_test_indices = sentences_to_indices(X_test, word_to_index, maxLen)
# Y_test_oh = np.eye(answer.shape[0]+1)[Y_test.reshape(-1)]

model = LSTM_model((maxLen,), word_to_vec_map, word_to_index, C = answer.shape[0]+1)
# model.summary()
model.compile(loss='categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
model.fit(X_train_indices, Y_train_oh, epochs = 200, batch_size = 16, shuffle = True)
```

```python
print("小刺郎: Hello! What is your name?")
user_name = input()
print("小刺郎: Hi " + user_name + ". I am here to answer your questions about grad applications. If you want to exit, simply enter Bye")

while(True):
    print(user_name + ": ", end="")
    user_response = input()
   
    if("bye" in user_response.lower().split()):
        print("小刺郎: Hope my answer has helped you. Bye!")
        break
    else:
        user_response_indices = sentences_to_indices(np.array([user_response]), word_to_index, maxLen)
        pred = np.argmax(model.predict(user_response_indices),axis = 1)
        bot_response = answer[pred[0]]
        print("小刺郎: ", end="")
        print(bot_response)
        print("小刺郎: What else can I help?")
     
```
